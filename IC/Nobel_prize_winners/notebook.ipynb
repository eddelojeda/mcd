### Analyze Nobel Prize winner data and identify patterns by answering the following questions:

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np
```

### What is the most commonly awarded gender and birth country?
- Store your answers as string variables `top_gender` and `top_country`.

```python
# Create DataFrame from nobel.csv 
df = pd.read_csv('data/nobel.csv')
# Finding the most commonly awarded gender and birth country
top_gender = df['sex'].value_counts().index[0]
top_country = df['birth_country'].value_counts().index[0]
print('Most commonly awarded gender:',top_gender,'\nMost commonly awarded birth country:', top_country)
```

### Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?
- Store this as an integer called `max_decade_usa`.

```python
# Finding which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories
# Adding the column with the first year of the corresponding decade
df['decade'] = (np.floor(df['year'] / 10) * 10).astype(int)
# Finding the number of winners and us winners
winners = df.groupby(['decade'])['year'].count()
us_winners = df[df.birth_country == 'United States of America'].groupby(['decade']).year.count()
# Porportion of us winners to total winners
prop_us_winners = (us_winners / winners).reset_index()
max_decade_usa = prop_us_winners[prop_us_winners.year == max(prop_us_winners.year)]['decade'].item()
print('Decade with the highest ratio of US-born Nobel Prize winners to total winners:',max_decade_usa,'-',max_decade_usa+9)
```

### Which decade and Nobel Prize category combination had the highest proportion of female laureates?
- Store this as a dictionary called `max_female_dict` where the decade is the key and the category is the value. There should only be one key:value pair.

```python
# Finding which decade and Nobel Prize category combination had the highest proportion of female laureates
# Adding the column that indicates if the winner was a female
df['female_winner'] = df['sex'] == 'Female'
# Porportion of female winners to total winners
prop_female_winners = df.groupby(['decade', 'category'], as_index=False)['female_winner'].mean()
max_decade_category_female = prop_female_winners[prop_female_winners.female_winner == max(prop_female_winners.female_winner)][['decade', 'category']]
dict_decade = max_decade_category_female['decade'].item()
dict_category = max_decade_category_female['category'].item()
max_female_dict = {dict_decade: dict_category}
print('Decade and Nobel Prize category combination had the highest proportion of female laureates:\nDecade:',dict_decade,'\nCategory:dict_category')
```

### Who was the first woman to receive a Nobel Prize, and in what category?
- Save your string answers as `first_woman_name` and `first_woman_category`.

```python
# Finding the first woman to receive a Nobel Prize
female_winners = df[df.sex == 'Female'].reset_index()
first_woman_name = female_winners['full_name'][0]
first_woman_category = female_winners['category'][0]
print('First woman to receive a Nobel Prize:',first_woman_name,'\nCategory:',first_woman_category)
```

### Which individuals or organizations have won more than one Nobel Prize throughout the years?
- Store the full names in a list named `repeat_list`.

```python
# Finding which individuals or organizations have won more than one Nobel Prize
winners_name = df.groupby(['full_name'], as_index=False)['year'].count()
repeat_list = winners_name[winners_name.year > 1 ]['full_name'].tolist()
print('Individuals or organizations have won more than one Nobel Prize:')
for i in range(len(repeat_list)):
    print('-',repeat_list[i])
```
